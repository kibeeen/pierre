```python
import pandas as pd
import sqlite3 as db
import os

# Store project file path for easy access
pwd = os.getcwd()
```

```python
df = pd.read_excel(pwd + '\\Data - Homestead Data Skill Test.xlsx', sheet_name="data")
```

```python

pivot_table = pd.pivot_table(
    df,
    values=['Spend', 'Attributed Rev (1d)', 'Imprs', 'Visits', 'New Visits', 'Transactions (1d)','Email Signups (1d)'],
    index=['Platform (Northbeam)'],
    aggfunc=sum,
    margins=True,
    margins_name='Grand Total'
)


# Add 'Grand Total' row to bottom of pivot table
# pivot_table.loc['Grand Total'] = pivot_table.sum()
grand_total = pivot_table.loc['Grand Total']
pivot_table = pivot_table.drop('Grand Total')
pivot_table = pivot_table.append(grand_total)
# pivot_table.loc['Grand Total'] = pivot_table.sum()
print(pivot_table)



# Sort by revenue descending
pivot_table = pivot_table.reindex(columns=['Spend', 'Attributed Rev (1d)', 'Imprs', 'Visits',  'New Visits', 'Transactions (1d)','Email Signups (1d)'])
pivot_table = pivot_table.sort_values(['Attributed Rev (1d)', 'Transactions (1d)'], ascending=False)
pivot_table = pivot_table.round(2)




print(pivot_table)



# Print the pivot table
# print(pivot_table)
```
