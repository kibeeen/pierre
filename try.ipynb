```python
conn = db.connect('test.db')
cursor = conn.cursor()


#getting column length (61)
total_employees = dataset.columns.size

# Return the first value of the `Attendance Data` column
date_of_attendance = dataset['Attendance Date'].values[0]
date_of_attendance = pd.to_datetime(dataset['Attendance Date'])



# Return the total days from the `date_of_attendance` (30)
# total_days_in_month =  pd.Period(date_of_attendance).dt.daysinmonth[0]
total_days_in_month = date_of_attendance.dt.days_in_month[0]
# print(total_days_in_month)



# This is me transitioning from Frontend Web Engineer into a Data Engineer, right now this is the best way I can think of to insert the dataset into the database. I used for loop to extract the needed data from this particular excel format.
for i in range(0, total_days_in_month, 1):
  for j in range(1, total_employees, 1):

    attendance_date = dataset['Attendance Date'].iloc[i].strftime('%Y-%m-%d')
    # print (dataset['Attendance Date'].iloc[i])

    attendance_status = dataset.loc[i].values[j]
    # print (dataset.loc[i].values[j])

    employee_name = dataset.columns[j]
    # print(i,' ',dataset['Attendance Date'].iloc[i], ' ', dataset.columns[j], ' ', dataset.loc[i].values[j])

    data = (attendance_date, attendance_status, employee_name)
    insert_sql = "INSERT INTO employee_attendance(attendance_date, attendance_status, employee_name) VALUES (?, ?, ?)"

    cursor.execute(insert_sql, data)

conn.commit()
conn.close()
```
